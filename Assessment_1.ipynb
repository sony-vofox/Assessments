#### 3.	Explain the concepts of vectorization and broadcasting in NumPy arrays with examples.

### Vectorization

Vectorization is the process of applying the same process to all elements in an array. It can be used to replace the for loop.
    

```python
# For example

x = np.array([1,2,3,4])
num = 10
y = x+num
```

```python
y
```

Here, the number '10' is added to each element in an array because of Vectorization.

### Broadcasting
Broadcasting is used to perform operations on different dimensional arrays.
The broadcasting can be only be applied if the axes of 2 array match or either of them is 1. 
eg : If 'a' has a dimension of (3,1,5) and 'b' has a dimension of (1,4,5), then they can broadcast.
* Beacuse, axis 0 ---> 3 and 1, so beacuse of 1 they are eligible for broadcasting.
* In axis 1 --> 1 and 4, because of 1 they can broadcast.
* In axis 2, ---> 5 and 5 are matching, so they can broadcast.
