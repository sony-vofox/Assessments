Box plot clearly shows a average mean of age which comes in each Pclass category.
* Average age of Pclass 1 is around 37.
* Average age of Pclass 2 is around 29. 
* Average age of Pclass 3 is around 25.

So we can replace the missing values of 'Age' with these values.

```python
def fill_age(df):
    df['Age'] = df['Age'].fillna(df.groupby('Pclass')['Age'].transform('mean'))
    return df
df = fill_age(df)
df.tail()
```

```python
# Checking for Missing values

sns.heatmap(df.isnull(), yticklabels=False)
```

```python
# As 'Cabin' column has 77% of missing data's.. 
# Filling the missing values with 'Unknown' data. 

df['Cabin'] = df['Cabin'].fillna('Unknown')
```

### Age

```python
sns.boxplot(data = df, y=df['Age'])
```

```python
# Logarithemic Transformation

# df['Age'] = np.log1p(df['Age'])
```

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot

import scipy.stats as stat
import pylab
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Age')
```

```python
# Logarithemic Transformation

df['Age_log'] =np.log(df['Age'])
plot_data(df,'Age_log')
```

```python
# Square Root Transformation

df['Age_sqaure']=df.Age**(1/2)
plot_data(df,'Age_sqaure')
```

```python
#Exponential Transformation

df['Age_exponential'] = df.Age**(1/1.2)
plot_data(df,'Age_exponential')
```

```python
df['Age_boxcox'],parameters = stat.boxcox(df['Age'])
print(parameters)
plot_data(df,'Age_boxcox')
```

* So we can use either Exponentially Transformed 'Age' Column or Boxcox transformed 'Age' column

```python
# sns.boxplot(data = df, y=df['Age_boxcox'])

plt.figure(figsize=(14,6))
plt.subplot(1,3,1)
sns.boxplot(data = df, y=df['Age'])
plt.subplot(1,3,2)
sns.boxplot(data = df, y=df['Age_boxcox'])
plt.subplot(1,3,3)
sns.boxplot(data = df, y=df['Age_exponential'])
plt.show()
```
